```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
import cv2 

from sklearn.cluster import KMeans
```

```python
image = mpl.image.imread("meowMeow.webp")
```

```python
plt.imshow(image)
```

```python
image.shape
```

### **RESHAPING**

flatten two dimensions 

```python
X = image.reshape(-1, 3)
```

```python
X.shape
```

### **Training the MODEL**

 more the number of clusters the more detailed image will be

```python
km = KMeans(n_clusters = 4)
```

```python
km.fit(X)
```

```python
segmentedImg = km.cluster_centers_[km.labels_]
segmentedImg = segmentedImg.reshape(image.shape)
```

```python
plt.imshow(segmentedImg / 255)
```

## Exporting the image

cv2 uses BGR (blue green red) we will change into RGB 

```python
cv2.imwrite("segmentedImg.png" , cv2.cvtColor(segmentedImg.astype("uint8") , cv2.COLOR_BGR2RGB))
```
